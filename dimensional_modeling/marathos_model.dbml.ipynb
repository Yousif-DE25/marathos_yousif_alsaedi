```python
// Marathos - Dimensional Model

Table fct_results {
  result_id     long        [primary key, note: "Unik rad-id"]
  event_id      integer     [ref: > dim_event.event_id,   note: "FK till dim_event"]
  athlete_id    integer     [ref: > dim_athlete.athlete_id, note: "FK till dim_athlete"]
  performance_seconds integer [note: "Tid i sekunder – för km/mi-events"]
  performance_km      float   [note: "Distans i km – för h-events"]
  athlete_avg_speed   float   [note: "Genomsnittlig hastighet"]
}

Table dim_event {
  event_id              integer [primary key]
  event_name            string  [note: "Namn på eventet"]
  event_distance_length string  [note: "T.ex. 50km, 100mi, 24h"]
  event_unit            string  [note: "km, mi eller h"]
  event_num_finishers   integer [note: "Antal som fullföljde"]
  year_of_event         integer [note: "År eventet hölls"]
  event_dates           string  [note: "Datum för eventet"]
}

Table dim_athlete {
  athlete_id          integer [primary key]
  athlete_country     string  [note: "Landkod t.ex. SWE, USA"]
  athlete_gender      string  [note: "M eller F"]
  athlete_year_of_birth integer [note: "Födelseår"]
  athlete_approx_age  integer [note: "Ungefärlig ålder vid loppet"]
  athlete_age_category string  [note: "Ålderskategori t.ex. M45"]
}

```
