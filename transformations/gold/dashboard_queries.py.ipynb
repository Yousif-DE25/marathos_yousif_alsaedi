```python
# Kör dessa queries för att skapa visualiseringar i Databricks dashboard

# Metric 1: Totalt antal lopp
spark.sql(f"""
    SELECT count(*) as total_lopp
    FROM {SCHEMAS['gold']}.fct_results
""").display()

# Metric 2: Totalt antal unika löpare
spark.sql(f"""
    SELECT count(DISTINCT athlete_id) as unika_lopare
    FROM {SCHEMAS['gold']}.fct_results
""").display()

# Metric 3: Antal länder representerade
spark.sql(f"""
    SELECT count(DISTINCT athlete_country) as antal_lander
    FROM {SCHEMAS['gold']}.dim_athlete
""").display()

# Graf 1: Top 10 länder per antal löpare
spark.sql(f"""
    SELECT 
        a.athlete_country,
        count(DISTINCT f.athlete_id) as antal_lopare
    FROM {SCHEMAS['gold']}.fct_results f
    JOIN {SCHEMAS['gold']}.dim_athlete a ON f.athlete_id = a.athlete_id
    GROUP BY a.athlete_country
    ORDER BY antal_lopare DESC
    LIMIT 10
""").display()

# Graf 2: Åldersfördelning bland löpare
spark.sql(f"""
    SELECT 
        athlete_approx_age as alder,
        count(*) as antal
    FROM {SCHEMAS['gold']}.dim_athlete
    WHERE athlete_approx_age IS NOT NULL
    GROUP BY alder
    ORDER BY alder
""").display()

# Graf 3: Antal lopp per år
spark.sql(f"""
    SELECT 
        e.year_of_event,
        count(*) as antal_lopp
    FROM {SCHEMAS['gold']}.fct_results f
    JOIN {SCHEMAS['gold']}.dim_event e ON f.event_id = e.event_id
    WHERE e.year_of_event IS NOT NULL
    GROUP BY e.year_of_event
    ORDER BY e.year_of_event
""").display()

# Graf 4: Fördelning av event-typer
spark.sql(f"""
    SELECT 
        e.event_unit,
        count(*) as antal_lopp
    FROM {SCHEMAS['gold']}.fct_results f
    JOIN {SCHEMAS['gold']}.dim_event e ON f.event_id = e.event_id
    GROUP BY e.event_unit
    ORDER BY antal_lopp DESC
""").display()

# Graf 5: Genomsnittlig hastighet per kön
spark.sql(f"""
    SELECT 
        a.athlete_gender as kon,
        round(avg(try_cast(f.athlete_avg_speed as float)), 2) as snitt_hastighet
    FROM {SCHEMAS['gold']}.fct_results f
    JOIN {SCHEMAS['gold']}.dim_athlete a ON f.athlete_id = a.athlete_id
    WHERE f.athlete_avg_speed IS NOT NULL
    GROUP BY a.athlete_gender
    ORDER BY snitt_hastighet DESC
""").display()
```
