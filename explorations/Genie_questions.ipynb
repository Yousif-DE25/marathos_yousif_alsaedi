```python
# Manuella SQL-queries för att dubbelkolla att Genie ger korrekta svar

# Fråga 1: Vilket land har flest löpare?
# Genie svarade: USA med 270,903 löpare
spark.sql(f"""
    SELECT 
        a.athlete_country,
        count(DISTINCT f.athlete_id) as antal_lopare
    FROM {SCHEMAS['gold']}.fct_results f
    JOIN {SCHEMAS['gold']}.dim_athlete a ON f.athlete_id = a.athlete_id
    GROUP BY a.athlete_country
    ORDER BY antal_lopare DESC
    LIMIT 5
""").display()

# Fråga 2: Hur många lopp hölls år 2019?
# Genie svarade: 7,611 lopp
spark.sql(f"""
    SELECT 
        count(DISTINCT e.event_id) as antal_unika_lopp
    FROM {SCHEMAS['gold']}.fct_results f
    JOIN {SCHEMAS['gold']}.dim_event e ON f.event_id = e.event_id
    WHERE e.year_of_event = 2019
""").display()

# Fråga 3: Vad är genomsnittsåldern bland löpare?
# Genie svarade: 41.6 år
spark.sql(f"""
    SELECT 
        round(avg(athlete_approx_age), 1) as snitt_alder,
        min(athlete_approx_age) as min_alder,
        max(athlete_approx_age) as max_alder,
        count(DISTINCT athlete_id) as antal_lopare
    FROM {SCHEMAS['gold']}.dim_athlete
    WHERE athlete_approx_age IS NOT NULL
""").display()
```
